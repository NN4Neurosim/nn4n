## Define a Single-Area RNN

```python
params = {
    "n_areas": 1,
    "area_connectivities": np.array([[0.5]]), # define sparsity
    "hidden_size": 100,
    "input_size": 2,
    "output_size": 2,
}

ma = MultiArea(**params)
ma.visualize()
```

## Define a Multi-Area RNN

```python
area_connectivities = np.array([
    [1.0, 0.1, 0.02],
    [0.1, 1.0, 0.1],
    [0.02, 0.1, 1.0]
])
n_areas = [10, 50, 40]
params = {
    "n_areas": n_areas,
    "area_connectivities": area_connectivities,
    "hidden_size": 100,
    "input_size": 2,
    "output_size": 2,
    "input_areas": [0],
    "output_areas": [1],
}

ma = MultiArea(**params)
ma.visualize()
```

## Define a EI constrainted Multi-Area RNN

```python
params = {
    "n_areas": 3,
    "area_connectivities": [0.1, 0.05],
    "hidden_size": 90,
    "input_size": 2,
    "output_size": 2,
    "input_areas": [0],
    "output_areas": [1],
}

ma = MultiAreaEI(**params)
ma.visualize()
```

```python
area_connectivities = np.array([
    [1.0, 0.05, 0.0],
    [0.1, 1.0, 0.05],
    [0.0, 0.1, 1.0]
])
params = {
    "n_areas": 3,
    "area_connectivities": area_connectivities,
    "hidden_size": 120,
    "input_size": 2,
    "output_size": 2,
    "input_areas": [0],
    "output_areas": [1],
}

ma = MultiAreaEI(**params)
ma.visualize()
```

## Define a EI constrainted Multi-Area RNN, but only excitatory neurons have intra-area connections

```python
area_connectivities = np.array([
    [1.0, 0.05, 0.0],
    [0.1, 1.0, 0.05],
    [0.0, 0.1, 1.0]
])
params = {
    "n_areas": 3,
    "area_connectivities": area_connectivities,
    "hidden_size": 120,
    "input_size": 2,
    "output_size": 2,
    "input_areas": [0],
    "output_areas": [1],
    "intra_area_connections": [True, False, False, False],
}

ma = MultiAreaEI(**params)
ma.visualize()
```
